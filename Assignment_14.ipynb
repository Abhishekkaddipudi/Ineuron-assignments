<h2>1. What does RGBA stand for?</h2>

- RGBA stands for "Red Green Blue Alpha". It is a color model used to represent colors in digital images and computer graphics. 


- The "Red", "Green", and "Blue" components represent the amounts of red, green, and blue light that combine to create a color, while the "Alpha" component represents the level of transparency or opacity of the color

<h2>2. From the Pillow module, how do you get the RGBA value of any images?</h2>

```python
from PIL import Image
image = Image.open("example.png")
x,y=(25,25)
rgba = image.getpixel((x, y))
display(image)
print(rgba)
```

<h2>3. What is a box tuple, and how does it work?</h2>

 - box tuple is a tuple of four values that represent the position and size of a rectangular area.

- a box tuple is commonly used to represent regions of an image or an area where an operation will be applied.

```python
from PIL import Image
image = Image.open("example.png")
display(image)
box = (185,90,495,305)
cropped_image = image.crop(box)
display(cropped_image)
```

<h2>4. Use your image and load in notebook then, How can you find out the width and height of an Image object?</h2>

> To find out the width and height of an Image object in Python using the Pillow module, we can use the size attribute of the Image object.

```python
image = Image.open("example.png")
width, height = image.size
display(image)
print("Width:", width)
print("Height:", height)
```

<h2>6. After making changes to an Image object, how could you save it as an image file?</h2>

> To save an Image object as an image file in Python using the Pillow module, you can use the save() method of the Image object

```python
from PIL import Image 
image = Image.open("example.png")
image = image.convert('1') # convert image to black and white
image.save('result.png')
display(Image.open("result.png"))# note saved image is displayed
```

<h2>7. What module contains Pillow’s shape-drawing code?</h2>

- Pillow's shape-drawing code is included in the ImageDraw module of the Pillow library. 



- The ImageDraw module provides a Draw object that can be used to draw various shapes

```python
from PIL import Image, ImageDraw
image = Image.open("example.png")
draw = ImageDraw.Draw(image)
draw.rectangle((0, 250, height, height), fill=(0, 0, 0))
display(image)
```

<h2>8. Image objects do not have drawing methods. What kind of object does? How do you get this kind of object?</h2>

- the Draw object in the ImageDraw module have drawing methods.



- we can draw shapes, lines, and text on an Image object using the Draw object

To create a Draw object, you can call the ImageDraw.Draw() method and pass in the Image object that you want to draw on.

```python
from PIL import Image, ImageDraw
image = Image.open("example.png")
draw = ImageDraw.Draw(image)
```

this creates 'draw' object which can be used to draw shapes,lines and text on the image

```python
draw.rectangle((50, 50, 150, 150), fill=(0, 0, 255))
draw.line((0,0,width,height))
draw.text((width/2,height/2),"this is a text")
display(image)
```
